### Solution

Step 1
To calculate the average scores between rotten_tomatoes and rotten_tomatoes_user, you can add them together and then divide them by 2.

```r
rotten_tomatoes_avg <- (rotten_tomatoes + rotten_tomatoes_user)/2
```

Step 2
To create a data frame with all of the information of interest, use the data.frame() function.

```r
df <- data.frame(film, rotten_tomatoes, rotten_tomatoes_user, rotten_tomatoes_avg)
```

Step 3
To sort a data frame based on a category, use the order() function. By default, the system will sort the data in ascending order, to sort in descending order, place a minus sign - in front of the category.

```r
sorted <- df[order(-rotten_tomatoes_avg),]
```

Step 4
To extract just the top ten rows of information, you can use the head() function and specify n = 10

```r
top_10 <- head(sorted, n = 10)
```

```r
# write your code below in RStudio
# then press the Source button to run it

suppressPackageStartupMessages(library(tidyverse))
library(tidyverse)
theme_update(plot.title = element_text(hjust = 0.5))

# add your code below this line

d <- read.csv("data/fandango.csv")

film <- d$FILM
rotten_tomatoes <- d$RottenTomatoes
rotten_tomatoes_user <- d$RottenTomatoes_User

rotten_tomatoes_avg <- (rotten_tomatoes + rotten_tomatoes_user)/2

df <- data.frame(film, rotten_tomatoes, rotten_tomatoes_user, rotten_tomatoes_avg)

sorted <- df[order(-rotten_tomatoes_avg),]

top_10 <- head(sorted, n = 10)

top_10_film <- rep(top_10$film, each = 3)
category <- rep(c("rotten_tomatoes", 
                  "rotten_tomatoes_user", 
                  "rotten_tomatoes_avg"), 10)
top_10_score <- top_10[-1]
score <- as.vector(t(top_10_score))
df2 <- data.frame(top_10_film, category, score)

chart <- ggplot(df2, aes(x = reorder(top_10_film, score), y = score, fill = category)) +
  geom_bar(position = "stack", stat = "identity") +
  labs(title = "Highest Averaged Scoring Movies\nby Rotten Tomatoes and Users\nin 2014-2015",
       x = "Film",
       y = "Scores",
       fill = "Category") +
  geom_text(aes(label = score), size = 4, position = position_stack(vjust = 0.5)) +
  coord_flip()
print(chart)

# add your code above this line

# DO NOT EDIT CODE BELOW THIS LINE

require(png)
png("png/lab/chart.png", width = 935, height = 455)
plot(chart)
invisible(dev.off())

print(chart)

img1 <- readPNG("png/lab/chart.png")
img2 <- readPNG("png/lab/lab.png")

if((100*sum(img1==img2)/length(img1)) >= 80) {
  print(TRUE)
} else {
  print(FALSE)
}

# DO NOT EDIT CODE ABOVE THIS LINE
```
