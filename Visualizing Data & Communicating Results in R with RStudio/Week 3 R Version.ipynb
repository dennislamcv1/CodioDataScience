```r
# Some sector and geographies for illustration
column <- c("Col 1","Col 2", "Col 3", "Col 4")
row <- c("Row 1", "Row 2", "Row 3", "Row 4")
# Simulate market returns based on a random normal distribution, with a mean return of 8% and volatility (standard deviation) 15%
numbers <- data.frame(Returns = round(rnorm(n = 1000, mean = 0.08, sd = 0.15), 2)) 

mydata <- cbind(Column = sample(column, 1000, replace = T), Row = sample(row, 1000, replace = T), numbers)
mydata
```

```r
plot <- ggplot(mydata, aes(x = Column, Row)) +
  geom_tile(aes(fill = Returns))
print(plot)
```
